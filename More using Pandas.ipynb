```python
titanic_data['name words'] = titanic_data.Name.str.split(',')
```

```python
titanic_data['first name'] = titanic_data['name words'].map(lambda fn : fn[0])
```

```python
titanic_data['last name'] = titanic_data['name words'].map(lambda fn : fn[-1])
```
